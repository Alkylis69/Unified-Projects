# Wine Quality Dataset Study

#### Motive:
- Studying the acids present in the wine data collected in the .csv file.
- Plotting the distribution of the acids.
- Calculating the central tendencies of the acids

## Insights

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Wine Quality Dataset.csv")
df.head(
)
```

```python
df.rename(columns={df.columns[2]:'citric acidity'},inplace =True)
df.head()
```

```python
df.info()
```

## Plotting the distribution of each acidity parameter

Plot **histograms**, check for **skewness**, find the **mean** and **median**

### 1. Fixed acidity:

```python
# Calculating the mean and median of fixed acidity
fa_mean = df['fixed acidity'].mean()
fa_median = df['fixed acidity'].median()

print(f'Mean: {fa_mean:00.2f},\nMedian: {fa_median}')
```

```python
# Plotting the histogram for fixed acidity

plt.figure(figsize=(13,9))

sns.histplot(data=df, x='fixed acidity',color='#f0c348',alpha=0.8,kde=True,)

plt.title('Fixed Acidity Distribution', fontdict={'size':22})

# Plotting the vertical lines for mean and median
plt.vlines(fa_mean, ymin=0,ymax=550,color='cyan',label=f'mean: {fa_mean:0.2f}')
plt.vlines(fa_median, ymin=0,ymax=550,color='red',label=f'median: {fa_median:0.2f}')

plt.legend()

plt.show()
```

```python
# Finding the Skewness

fa_skew=df['fixed acidity'].skew()
print(f'Skewness of fixed acidity distribution: {fa_skew:.2f}')
```

### 2. Volatile Acidity

```python
# Calculating the mean and median of volatile acidity
va_mean=df['volatile acidity'].mean()
va_median=df['volatile acidity'].median()

print(f'Mean: {va_mean:0.2f}\nMedian: {va_median:0.2f}')
```

```python
# Plotting the histogram for volatile acidity

plt.figure(figsize=(13,9))

sns.histplot(data=df, x='volatile acidity',color='#d595fc',alpha=0.8,kde=True)

plt.title('Volatile Acidity Distribution', fontdict={'size':22})

# Plotting the vertical lines for mean and median
plt.vlines(va_mean, ymin=0,ymax=500,color='#ffdd00',label=f'mean: {va_mean:0.2f}')
plt.vlines(va_median, ymin=0,ymax=500,color='#07fa07',label=f'median: {va_median:0.2f}')

plt.legend()

plt.show()
```

```python
# Finding the Skewness

va_skew=df['volatile acidity'].skew()
print(f'Skewness of volatile acidity distribution: {va_skew:.2f}')
```

### 3. Citric Acidity

```python
# Calculating the mean and median of citric acidity
ca_mean=(df['citric acidity'].mean())
ca_median=df['citric acidity'].median()

print(f'Mean: {ca_mean:0.2f}\nMedian: {ca_median:0.2f}')
```

```python
# Plotting the histogram for volatile acidity

plt.figure(figsize=(13,9))

sns.histplot(data=df, x='citric acidity',color='#6200ff',alpha=0.3,kde=True)

plt.title('Citric Acidity Distribution', fontdict={'size':22})

# Plotting the vertical lines for mean and median
plt.vlines(ca_mean, ymin=0,ymax=500,color='#ffdd00',label=f'mean: {ca_mean:0.2f}')
plt.vlines(ca_median, ymin=0,ymax=500,color='#ff7300',label=f'median: {ca_median:0.2f}')

plt.legend()

plt.show()
```

```python
# Finding the Skewness

ca_skew=df['citric acidity'].skew()
print(f'Skewness of citric acidity distribution: {ca_skew:.2f}')
```

### 4. Quality

Next, we measure the quality index of the wines... Since its a discrete distribution, we obtain the central tendency through **Mode**.

```python
plt.figure(figsize=(11,6))

sns.countplot(data=df, x='quality',hue='quality')

plt.xlabel('Quality')
plt.ylabel('Count')
plt.title('Count plot of Quality Index')

plt.show()
```

#### Obtaining the mode of the quality index

```python
df['quality'].value_counts()
```

```python
# Mode is the most occuring value in the dataset 

mode_quality=df['quality'].value_counts().index[0]
print(f'Mode for the quality index if the wines: {mode_quality}')
```

## Allocating Optimal Values for the Acids
Choosing the appropriate _central tendency_ by studying the distribution plots

```python
l=df.columns[:3]
l=list(l)
l.append(df.columns[11])
l
```

_For positively skewed distributions, **Median** is an appropriate measure_

```python
optimum_wines=pd.Series(
    index=l,
    data= [fa_mean,va_median,ca_median,mode_quality]
)
print ('Optimal Acidity for the wines:\n')
print(optimum_wines)
```
